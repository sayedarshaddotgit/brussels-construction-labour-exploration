```python
municipalities = canal_zone[
    ['cd_munty_refnis','tx_munty_descr_fr','geometry']
].copy()

municipalities = municipalities.dissolve(
    by='cd_munty_refnis'
).reset_index()

print(municipalities.shape)

municipalities[['cd_munty_refnis']].head()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

municipalities.plot(
    ax=ax,
    edgecolor='black',
    linewidth=1
)

for idx, row in municipalities.iterrows():
    centroid = row.geometry.centroid
    ax.annotate(
        row['cd_munty_refnis'],
        (centroid.x, centroid.y),
        ha='center'
    )

ax.set_title("Canal Zone Municipalities")
ax.axis('off')

plt.show()
```
